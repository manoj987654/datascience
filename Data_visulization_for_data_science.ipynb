# **Data Visulization for Data Sience**

## **LINEPLOT**

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri=sns.load_dataset("fmri")
```

```python
fmri.head()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri)
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event')
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event')
plt.show()
```

```python
sns.lineplot(x='timepoint',y='signal',data=fmri,hue='event',style='event',markers=True)
plt.show()
```

## **BARPLOT**

```python
import pandas as pd
data=pd.read_csv('pokemon.csv')
data.head()
```

```python
sns.barplot(x='Type 2',y='Type 1',data=data)
plt.show()
```

```python
data.columns
```

```python
sns.barplot(x='Legendary',y='Speed',data=data,color='red')
plt.show()
```

```python
sns.barplot(x='Legendary',y='Speed',data=data,hue='Generation',palette='vlag')
plt.show()
```

```python
sns.barplot(x='Legendary',y='Speed',data=data,hue='Generation',ci=None)
plt.show()
```

## **SCATTER PLOT**

```python
iris=pd.read_csv('iris.csv')
```

```python
iris.head(20)
```

```python
iris.columns
```

```python
sns.scatterplot(x='SepalLengthCm',y='PetalLengthCm',data=iris,color='green')
plt.show()
```

```python
sns.scatterplot(x='SepalLengthCm',y='PetalLengthCm',data=iris,hue='Species',color='green')
plt.show()
```

```python
sns.scatterplot(x='SepalLengthCm',y='PetalLengthCm',data=iris,palette='vlag',hue='PetalLengthCm')
plt.show()
```

## **HISTOGRAM**

```python
diamond=pd.read_csv('diamond.csv')
diamond.head()
```

```python
sns.distplot(diamond['Depth'])
plt.show()
```

```python
sns.distplot(diamond['Depth'],hist=False)
plt.show()
```

```python
sns.distplot(diamond['Depth'],color='green')
plt.show()
```

```python
sns.distplot(diamond['Depth'],color='red',kde=False,bins=10)
plt.show()
```

```python
sns.distplot(diamond['Depth'],color='red',kde=False,bins=10,vertical=True)
plt.show()
```

## **BOXPLOT**

```python
sns.boxplot(x='SepalLengthCm',y='PetalLengthCm',data=iris,color='red')
plt.show()
```

```python
sns.boxplot(x='SepalLengthCm',y='PetalLengthCm',data=iris,hue='Species')
plt.show()
```

```python
sns.boxplot(x='SepalLengthCm',y='PetalLengthCm',data=iris,hue='Species',notch=True)
plt.show()
```
